```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
bmi = [0, 25]
prob = [1, 0]

#polynomial fit with degree = 1
model = np.poly1d(np.polyfit(bmi, prob, 1))

#add fitted polynomial line to scatterplot
polyline = np.linspace(1, max(bmi), 50)
plt.scatter(bmi, prob, color= "#f16364")
plt.plot(polyline, model(polyline), color="black")
plt.title("Likelihood to befriend")
plt.xlabel("Difference in bmi")
plt.ylabel("Probability")
#plt.savefig("D:/juju/lumc/msc_thesis/report/images/growth_formula_" + name[i] +".png", format="png",  dpi=300, bbox_inches='tight')  
plt.show()
print(model)
```

```python
bmit = [0, 10]
probt = [1, 0]

#polynomial fit with degree = 1
modelt = np.poly1d(np.polyfit(bmit, probt, 1))

#add fitted polynomial line to scatterplot
polylinet = np.linspace(1, max(bmit), 50)
#plt.scatter(bmit, probt, color= "#f16364")
plt.plot(polylinet, modelt(polylinet), color="black", label="minimum acceptance")
plt.plot(polyline, model(polyline), color="#f16364", label="maximum acceptance")
plt.title("Likelihood to befriend")
plt.xlabel("Asbolute difference in bmi")
plt.ylabel("Probability")
plt.legend()
plt.savefig("D:/juju/lumc/msc_thesis/report/images/likelihood_to_befriend.png", format="png",  dpi=300, bbox_inches='tight')  
plt.show()
print(modelt)
```
