```python
data_dutch = {1: 75.0 , 2: 87.1 , 3: 97.0, 4: 104.9, 5: 112.1, 6: 118.8, 7: 125.3, 8: 131.3, 9: 137.3, 10: 143.5,
              11: 149.7, 12: 155.7, 13: 160.8, 14: 164.5, 15: 166.9, 16: 168.3, 17: 169.2, 18:169.7, 19:170.1, 20: 170.4, 21: 170.7}

height_turkish = [75.6, 88, 96, 102, 110, 117, 123, 128, 134, 140, 148, 153, 157, 160, 161, 161, 161, 161, 161, 161, 161]
height_moroccan = [75.0, 86.5, 96.0, 103.5, 110.2, 116.8, 123.0, 128.4, 134.1, 140.2, 147.0, 152.9, 157.3, 160.4, 162.4, 162.8, 162.8, 162.8, 162.8, 162.8, 162.8]
height_hindustan = [73.5, 85.8, 95.0, 102.8, 110.1, 116.7, 122.9, 129.1, 135.5, 141.7, 147.5, 152.4, 156.0, 158.2, 159.3, 159.6, 159.6, 159.6, 159.6, 159.6, 159.6] 
```

```python
month = np.array(list(data_dutch.keys())) * 12
height_dutch = list(data_dutch.values())
```

```python
name = ["Dutch", "Turkish", "Moroccan", " Hindustan"]

i = 0
for height in [height_dutch, height_turkish, height_moroccan, height_hindustan]:

    #polynomial fit with degree = 2
    model = np.poly1d(np.polyfit(month, height, 2))

    #add fitted polynomial line to scatterplot
    polyline = np.linspace(1, max(month), 50)
    plt.scatter(month, height, color= "#f16364")
    plt.plot(polyline, model(polyline), color="black")
    plt.title("Growth formula of " + name[i] + " women")
    plt.xlabel("Age (months)")
    plt.ylabel("Height (cm)")
    plt.savefig("D:/juju/lumc/msc_thesis/report/images/growth_formula_" + name[i] +".png", format="png",  dpi=300, bbox_inches='tight')  
    plt.show()
    i +=1
    print(model)
```
